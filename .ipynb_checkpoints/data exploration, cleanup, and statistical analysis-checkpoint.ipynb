# ILINet data Cleansing & Preparation

```python
## Pull in CSV file of 2013-2019 flu data
flu_csv = "Resources/ILINet2013-2019.csv"
raw_flu_df = pd.read_csv(flu_csv, encoding="utf-8")
raw_flu_df.describe()
```

```python
raw_flu_df.head()
```

```python
# Clean up the dataset by removing Florida..."x" for all occurrences
clean_flu_df = raw_flu_df[raw_flu_df['TOTAL PATIENTS'] != "X"]

# Removes 283 rows of unusable data
clean_flu_df.describe()
# clean_flu_df.dtypes
```

```python
# Function to find the associated flu season
# Flu season is defined by the majority of when the flu occurrence (in weeks) take place
# Flu season period is defined as: Q4 20XX - Q3 20XX+1 ...so for example, 10/1/17 - 9/30/2018
def fluSeason(row):
    if row['WEEK'] > 39:
        return row['YEAR'] + 1
    else:
        return row['YEAR'] 

# Function to assign week of flu season in order of 1 to 53 (since 40 technically is the first week)...
# Purpose is for use in Matplotlib representations
def fluWeek(row):
    order = list(range(1,54))
    a = list(range(40,54))
    b = list(range(1,40))
    a.extend(b)
    fluWkDict = dict(zip(a, order))

    for x, y in fluWkDict.items():
        if row['WEEK'] == x:
            return y

# Run the above functions with the flu dataframe
clean_flu_df['Flu_Season'] = clean_flu_df.apply(lambda x: fluSeason(x), axis=1)
clean_flu_df['Flu_Week'] = clean_flu_df.apply(lambda x: fluWeek(x), axis=1)

# Format the newly created flu week column to a whole number
# clean_flu_df['Flu_Week'] = clean_flu_df['Flu_Week'].map('{:,.0f}'.format)

# Convert other columns to integer
clean_flu_df['Flu_Week'] = clean_flu_df['Flu_Week'].astype(int)
clean_flu_df['ILITOTAL'] = clean_flu_df['ILITOTAL'].astype(int)
clean_flu_df['%UNWEIGHTED ILI'] = clean_flu_df['%UNWEIGHTED ILI'].astype(float)
clean_flu_df.head()


```

```python
# Checking datatypes
clean_flu_df.dtypes
```

```python
# Testing calculation for Flu Week and making sure there aren't any "nan" values - some years have additional "week" at the end
# 2015 had a "Week 53" at the end...had to take this into account when writing function for fluWeek above
clean_flu_df['Flu_Week'].value_counts()
```

```python
# Create clean state population dataframe
cc_pop = ccdf[['Name','Population']]
cc_pop.head()
```

```python
# Create new clean dataframe that contains only the columns that we need
clean2_flu_df = clean_flu_df[['REGION','Flu_Season','Flu_Week','YEAR','WEEK','ILITOTAL']]
# Add total US population for each state to calculate % ILITOTAL (in 100,000s)
clean2_flu_df = clean2_flu_df.merge(cc_pop,how='left',left_on='REGION',right_on='Name')
clean2_flu_df['%ILITOTAL'] = ((clean2_flu_df['ILITOTAL']/clean2_flu_df['Population'])*100000).astype(float)
clean2_flu_df = clean2_flu_df.drop('Name',axis=1)
clean2_flu_df = clean2_flu_df.drop('Population',axis=1)
clean2_flu_df.head()
```

```python
# Creating GroupBy dataframe for plotting yearly line data
grouped_flu_df = clean2_flu_df.groupby(by=['Flu_Season','Flu_Week'])['%ILITOTAL'].sum()
grouped_flu_df.reset_index().head()
```

```python
# Create national average by week (average 5 year data from all states) for line graph
fiveyr_natavg = clean2_flu_df.groupby(by=['Flu_Season','Flu_Week'])['%ILITOTAL'].sum()
fiveyr_natavg = fiveyr_natavg.reset_index()

nat_avg = fiveyr_natavg.groupby(by=['Flu_Week'])['%ILITOTAL'].mean()
nat_avg = nat_avg.reset_index()
nat_avg.head()
```

```python
State_ILI_df=census_flu_df2[['Name','Ratio']]
# Add columns for lat, lng
State_ILI_df['lat']=''
State_ILI_df['lng']=''
State_ILI_index=State_ILI_df.set_index('Name')

# Loop through to get the lat/long for each state
states=list(State_ILI_index.index)
for state in states:
    target_url = (f'https://maps.googleapis.com/maps/api/geocode/json?address={state}&key={g_key}')
    geo_data = requests.get(target_url).json()
    lat = geo_data["results"][0]["geometry"]["location"]["lat"]
    lng = geo_data["results"][0]["geometry"]["location"]["lng"]
    State_ILI_index.loc[state, 'lat'] = lat
    State_ILI_index.loc[state, 'lng'] = lng
```

```python
# flu case weighted by state population heatmap

# Store latitude and longitude in locations
locations = State_ILI_index[["lat", "lng"]]
ILIRatio = State_ILI_index["Ratio"]
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=ILIRatio, 
                                 dissipating=False, max_intensity=20,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# total flu case heatmap
total_flu=pd.merge(flu_five_years, State_ILI_index, how='inner', on='Name')

# Store latitude and longitude in locations
locations_total = total_flu[["lat", "lng"]]
ILI_total = total_flu["ILITOTAL"]/(total_flu["ILITOTAL"].sum())*100
# Plot Heatmap
fig_total = gmaps.figure()

# Create heat layer
heat_layer_total = gmaps.heatmap_layer(locations_total, weights=ILI_total, 
                                 dissipating=False, max_intensity=1,
                                 point_radius=1)


# Add layer
fig_total.add_layer(heat_layer_total)

# Display figure
fig_total
```

```python
from scipy.stats import spearmanr


# correlation with target
labels = []
values = []

target_col_name = "ILI Ratio (in 10,000)"

for col in census_flu_df5.columns:
    if col != target_col_name:
        labels.append(col)
        values.append(spearmanr(census_flu_df5[col].values, census_flu_df5[target_col_name].values)[0])

corr_df = pd.DataFrame({'col_labels': labels, 'corr_values' : values})
corr_df = corr_df.sort_values(by='corr_values')
### if you want to put a threshold on your correlation, do something like this:
# corr_df = corr_df[(corr_df['corr_values']>0.03) | (corr_df['corr_values']<-0.03)]

ind = np.arange(corr_df.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(12,12))
rects = ax.barh(ind, np.array(corr_df.corr_values.values), color='darkseagreen')
ax.set_yticks(ind)
ax.set_yticklabels(corr_df.col_labels.values, rotation='horizontal',fontsize=15)
ax.set_xlabel("Correlation coefficient",fontsize=20)
ax.set_title("Variable correlation to ILI Ratio",fontsize=22)
```

```python
# get the Latitude, poverty ratio and unemployment ratio, ILI ratio data
Cor_scat_df=census_flu_df5[['Latitude', 'ILI Ratio (in 10,000)', 'Poverty Ratio (%)',
        'Unemployment (%)']]
Cor_scat_df=Cor_scat_df.rename(columns={'Poverty Ratio (%)':'Poverty (%)','ILI Ratio (in 10,000)':'ILI number (in 10,000)'})
```

```python
# Linear regretion plot
import seaborn as sns
from scipy import stats

# get coeffs of linear fit, use line_kws to set line label for legend
slope1, intercept1, r_value, p_value, std_err = stats.linregress(Cor_scat_df['Poverty (%)'],Cor_scat_df['ILI number (in 10,000)'])
slope2, intercept2, r_value, p_value, std_err = stats.linregress(Cor_scat_df['Unemployment (%)'],Cor_scat_df['ILI number (in 10,000)'])
fig,axes=plt.subplots(1, 2, figsize=(10,5),sharey=True)
ax1=sns.regplot(x='Poverty (%)', y='ILI number (in 10,000)', data=Cor_scat_df,ax=axes[0],line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope1,intercept1)})
ax2=sns.regplot(x='Unemployment (%)', y='ILI number (in 10,000)', data=Cor_scat_df,ax=axes[1],line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope2,intercept2)})
plt.ylim(0,200)

ax1.legend()
ax2.legend()

```

# Flu-Hospitalization-Analysis

```python
# File to Load
file_to_load = "Resources/Final_FluSurveillance_FluSurv-NET_Entire Network_Data.csv"

# Read Purchasing File and store into Pandas data frame
Hos2017_2018_df = pd.read_csv(file_to_load)
Hos2017_2018_df.head()
```

```python
Hos2017_2018_df["MMWR-WEEK"]=Hos2017_2018_df["MMWR-WEEK"]+Hos2017_2018_df["MMWR-YEAR"] *52-2017*52
```

```python
Age_groups=Hos2017_2018_df.groupby(["AGE CATEGORY","MMWR-WEEK","SEASON"])["WEEKLY RATE "].sum()
Age=pd.DataFrame(Age_groups)
Age1=Age.reset_index()
```

```python
fig, ax = plt.subplots()

for key, grp in Age1.groupby(['AGE CATEGORY']):
    ax = grp.plot(ax=ax, kind='line', x='MMWR-WEEK', y='WEEKLY RATE ', label=key)

x_axis = list(range(40,70))
tick_locations = [value for value in x_axis]

# Tick locations are the same as number of weeks...so can use for location and labels
plt.xticks(tick_locations, tick_locations) 

# plt.xlim(40,92)
    
plt.legend(loc='best')
plt.title("Laboratory Comfirmed Flu Hospitalizations 2017-2018")
plt.xlabel("Flu Week")
plt.ylabel("Rates per 100,000 population")
plt.rcParams["figure.figsize"] = [20,20]
```

```python
# File to Load
file_to_load = "Resources/FluSurveillance_FluSurv-NET_ALLYears.csv"

# Read Purchasing File and store into Pandas data frame
HosALL_df = pd.read_csv(file_to_load)
```

```python
ALL_ages=HosALL_df.groupby(["AGE CATEGORY","MMWR-WEEK","SEASON"])["WEEKLY RATE "].sum()
All=pd.DataFrame(ALL_ages)
```

```python
Total=All.reset_index()
```

```python
Average1=Total[Total["AGE CATEGORY"]=="0-4 yr"]
Week1= Average1.groupby("MMWR-WEEK")["WEEKLY RATE "].mean()
Infant=pd.DataFrame(Week1)
Infant=Infant.rename(columns={"WEEKLY RATE ": "0-4 yr AVG"})
A=Infant.reset_index()
```

```python
Average2=Total[Total["AGE CATEGORY"]=="5-17 yr"]
Week2= Average2.groupby("MMWR-WEEK")["WEEKLY RATE "].mean()
Teens=pd.DataFrame(Week2)
Teens=Teens.rename(columns={"WEEKLY RATE ": "5-17 yr AVG"})
B=Teens.reset_index()
```

```python
Average3=Total[Total["AGE CATEGORY"]=="18-49 yr"]
Week3= Average3.groupby("MMWR-WEEK")["WEEKLY RATE "].mean()
Adults=pd.DataFrame(Week3)
Adults=Adults.rename(columns={"WEEKLY RATE ": "18-49 yr AVG"})
C=Adults.reset_index()
```

```python
Average4=Total[Total["AGE CATEGORY"]=="50-64 yr"]
Week4= Average4.groupby("MMWR-WEEK")["WEEKLY RATE "].mean()
Older=pd.DataFrame(Week4)
Older=Older.rename(columns={"WEEKLY RATE ": "50-64 yr AVG"})
D=Older.reset_index()
```

```python
Average5=Total[Total["AGE CATEGORY"]=="65+ yr"]
Week5= Average5.groupby("MMWR-WEEK")["WEEKLY RATE "].mean()
Old=pd.DataFrame(Week5)
Old=Old.rename(columns={"WEEKLY RATE ": "65+ yr AVG"})
E = Old.reset_index()
```

```python
merged_table= pd.merge(A, B, on="MMWR-WEEK", how="outer")
merged_table2=pd.merge(merged_table,C, on="MMWR-WEEK", how="outer")
merged_table3=pd.merge(merged_table2,D, on="MMWR-WEEK", how="outer")
merged_table4=pd.merge(merged_table3, E, on="MMWR-WEEK", how="outer")
merged_table4['Flu_Week'] = merged_table4['MMWR-WEEK'].apply(lambda x: (x+53 if x < 40 else x))
merged_table4.sort_values('Flu_Week', inplace=True)
```

```python
merged_table5 = merged_table4.set_index('Flu_Week').drop('MMWR-WEEK',axis=1)
```

```python
plt.rcParams["figure.figsize"] = [20,5]
merged_table5.plot(kind='line')
plt.legend(loc='best')
plt.title("Laboratory Comfirmed Flu Hospitalizations, Five Year Average")
plt.xlabel("Flu Week")
plt.ylabel("Rates per 100,000 population")
```
